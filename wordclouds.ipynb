```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL import Image
%matplotlib inline

num_rows = 100000
num_cols = 1000
data = {}
for i in range(num_cols):
    col_name = f'tejas{i+1}'
    data[col_name] = np.random.randint(low=0, high=100, size=num_rows)

#load data into a DataFrame object:
df = pd.DataFrame(data)


world = np.array(Image.open(r"Images/world.jpg"))
japan = np.array(Image.open(r"Images/japan.jpg"))
moradabad = np.array(Image.open(r"Images/moradabad.jpg"))
india = np.array(Image.open(r"Images/india.jpg"))
```

```python
# Creating a wordcloud of the words mostly found in reviews of the California branch.
plt.figure(figsize = (7,7))
Wc = WordCloud(mask =world, background_color='black',
               max_words = 1000 , width = 500 , height = 400, 
               contour_width = 1, contour_color = 'white', colormap = "viridis").generate(" ".join(df))
plt.axis("off")
plt.imshow(Wc , interpolation = 'bilinear')
```
